# Phase 1 — Data Loading & Inventory

**Well**: 15/9-F-1 A · **Field**: Volve, Norwegian North Sea  
**File**: `wells/15_9-F-1A.LAS`

This notebook covers the first phase of the log conditioning workflow:
- Load the LAS file and inspect the well header
- Inventory the available curve suite
- Assess data coverage (depth range and null fraction per curve)
- Display a standard 5-track composite log over the main logging interval

The composite display is the primary QC tool at this stage — it gives an immediate overview of data quality, lithology variation, and the depth ranges where each tool was run.

```python
import lasio
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from pathlib import Path

# ── Paths ───────────────────────────────────────────────────────────────────
WELL_FILE = Path('../wells/15_9-F-1A.LAS')

# ── Plot style ───────────────────────────────────────────────────────────────
plt.rcParams.update({
    'font.family'      : 'DejaVu Sans',
    'font.size'        : 9,
    'axes.linewidth'   : 0.8,
    'xtick.direction'  : 'in',
    'ytick.direction'  : 'in',
    'figure.dpi'       : 120,
})
```

## Step 1.1 — Load LAS File

`lasio` reads LAS 2.0 files and returns a `LASFile` object. The `.df()` method returns a `pandas.DataFrame` with depth as the index.

The null value in this file is `−999.25` (CWLS standard). We replace it with `NaN` immediately so that pandas statistics and matplotlib ignore null samples automatically.

```python
las = lasio.read(WELL_FILE)

# ── Well header ─────────────────────────────────────────────────────────────
header_items = [
    ('Well',     las.well['WELL'].value),
    ('Field',    las.well['FLD'].value),
    ('Company',  las.well['COMP'].value),
    ('Start (m)',las.well['STRT'].value),
    ('Stop (m)', las.well['STOP'].value),
    ('Step (m)', las.well['STEP'].value),
    ('Null',     las.well['NULL'].value),
    ('Location', las.well['LOC'].value),
    ('Lat',      las.well['LATI'].value),
    ('Long',     las.well['LONG'].value),
]
print('─' * 45)
print('  WELL HEADER')
print('─' * 45)
for k, v in header_items:
    print(f'  {k:<12s}: {v}')
print('─' * 45)
```

## Step 1.2 — Curve Inventory

Build the curve inventory matrix: mnemonic, units, and a plain-English description for each curve. This is the first deliverable of Phase 1 — it establishes what is available and what must be modelled or derived later.

```python
# ── Build curve inventory DataFrame ─────────────────────────────────────────
rows = []
for curve in las.curves:
    rows.append({
        'Mnemonic'   : curve.mnemonic,
        'Units'      : curve.unit,
        'Description': curve.descr,
    })

inventory = pd.DataFrame(rows)
print(f"Total curves loaded: {len(inventory)}")
print()
print(inventory.to_string(index=False))
```

## Step 1.3 — Data Coverage Assessment

For each curve, report:
- First and last depth where a valid (non-null) sample exists
- Number of valid samples and coverage percentage over the full file depth range

This immediately reveals which tools were run over which intervals, and flags any curves that are entirely absent or sparsely populated.

```python
# ── Load into DataFrame, replace nulls ──────────────────────────────────────
df = las.df()
df.replace(-999.25, np.nan, inplace=True)
df.index.name = 'DEPTH'

# ── Coverage table ───────────────────────────────────────────────────────────
total = len(df)
cov_rows = []
for col in df.columns:
    valid   = df[col].dropna()
    n_valid = len(valid)
    if n_valid > 0:
        z_top = valid.index[0]
        z_bot = valid.index[-1]
    else:
        z_top = z_bot = np.nan
    cov_rows.append({
        'Curve'    : col,
        'Top (m)'  : z_top,
        'Base (m)' : z_bot,
        'N valid'  : n_valid,
        'Coverage' : f"{100 * n_valid / total:.1f}%" if n_valid else '—',
        'Mean'     : f"{valid.mean():.3f}" if n_valid else '—',
        'Min'      : f"{valid.min():.3f}"  if n_valid else '—',
        'Max'      : f"{valid.max():.3f}"  if n_valid else '—',
    })

cov = pd.DataFrame(cov_rows)
print(cov.to_string(index=False))
```

## Step 1.4 — Rock Physics Curve Availability Check

Map the loaded curves against the minimum required suite for the log conditioning workflow. Curves that are absent will need to be derived (e.g. Vs from RPM, Vp from Faust) or sourced from external data.

```python
# ── Required curves and their roles ─────────────────────────────────────────
REQUIRED = {
    # Mnemonic : (Role, Critical?)
    'GR'    : ('Gamma ray — lithology indicator',         True),
    'CALI'  : ('Caliper — borehole QC',                   True),
    'BS'    : ('Bit size — caliper reference',            True),
    'RHOB'  : ('Bulk density — elastic input',            True),
    'DRHO'  : ('Density correction — RHOB quality',       True),
    'NPHI'  : ('Neutron porosity — lithology / density infill', True),
    'DT'    : ('Compressional sonic → Vp',               True),
    'DTS'   : ('Shear sonic → Vs (RPM calibration)',      True),
    'RT'    : ('True resistivity — Rw / saturation / Faust Vp', True),
    'PEF'   : ('Photoelectric factor — mineral ID',       False),
}

print(f"{'Curve':<8}  {'Present':<8}  {'Critical':<10}  Role")
print('─' * 70)
for mnem, (role, critical) in REQUIRED.items():
    present = mnem in df.columns and df[mnem].notna().any()
    flag    = '✓' if present else '✗  ← MISSING'
    crit    = 'Critical' if critical else 'Optional'
    print(f"{mnem:<8}  {flag:<8}  {crit:<10}  {role}")
```

## Standard Composite Log Display

The composite display is the standard petrophysical QC tool. It shows five tracks over a common depth axis:

| Track | Curves | Scale | Convention |
|-------|--------|-------|------------|
| 1 | GR (fill) + CALI + BS | 0–150 API · 6–16 in | Sand fill gold; shale fill green |
| 2 | RT resistivity | 0.2–2000 Ω·m (log) | Dark red; log scale critical |
| 3 | RHOB + NPHI overlay | 1.95–2.95 g/cc · 0.45–−0.15 v/v | Red/blue; NPHI reversed |
| 4 | DT + DTS sonic | 300–40 μs/ft (reversed) | Navy/crimson; fast rock → right |
| 5 | DRHO correction | −0.5–0.5 g/cc | Purple; ±0.15 threshold lines |

**Reading conventions**:
- Depth increases downward (y-axis inverted)
- On Track 1: GR < 60 API → sand (gold fill); GR > 60 → shale (green fill)
- On Track 3: when NPHI (blue) crosses to the left of RHOB (red) in the same porosity space → gas crossover
- On Track 4: DT and DTS diverge in gas sands (compressional slows more than shear)
- On Track 5: DRHO outside ±0.15 g/cc (red shading) flags unreliable density measurements

```python
def plot_composite_log(df, top, base, well_name='Well', bs_default=8.5, figsize=(16, 14)):
    """
    Standard 5-track composite well log display.

    Parameters
    ----------
    df          : DataFrame with DEPTH as index (nulls already set to NaN)
    top, base   : Depth interval to display (metres MD)
    well_name   : String for figure title
    bs_default  : Fallback bit size (inches) if BS curve is flat or absent
    figsize     : Figure size tuple
    """
    # ── Slice to display interval ────────────────────────────────────────────
    sub  = df.loc[top:base]
    depth = sub.index.values

    def get(col):
        """Return curve array or NaN array if curve is absent."""
        return sub[col].values if col in sub.columns else np.full(len(depth), np.nan)

    gr   = get('GR')
    cali = get('CALI')
    bs   = get('BS')
    rt   = get('RT')
    rhob = get('RHOB')
    nphi = get('NPHI')
    dt   = get('DT')
    dts  = get('DTS')
    drho = get('DRHO')

    # Use default bit size where BS is null or constant at surface value
    bs_plot = np.where(np.isnan(bs) | (bs > 30), bs_default, bs)

    # ── Figure layout — 5 tracks with unequal widths ─────────────────────────
    fig, axes = plt.subplots(
        1, 5, figsize=figsize, sharey=True,
        gridspec_kw={
            'width_ratios' : [1.4, 1.0, 1.4, 1.4, 0.9],
            'wspace'       : 0.06,
        }
    )

    # ─────────────────────────────────────────────────────────────────────────
    # Track 1 — Gamma Ray + Caliper
    # ─────────────────────────────────────────────────────────────────────────
    ax = axes[0]
    GR_SAND_CUT = 60  # API — adjust for local shale baseline

    ax.plot(gr, depth, color='#4A7C40', lw=0.9, label='GR')
    ax.fill_betweenx(depth, gr, GR_SAND_CUT,
                     where=(gr < GR_SAND_CUT),
                     color='#F5C518', alpha=0.55, label=f'Sand (GR<{GR_SAND_CUT})')
    ax.fill_betweenx(depth, GR_SAND_CUT, gr,
                     where=(gr > GR_SAND_CUT),
                     color='#7FB069', alpha=0.20, label='Shale')
    ax.set_xlim(0, 150)
    ax.xaxis.set_label_position('top')
    ax.xaxis.tick_top()
    ax.set_xlabel('GR (API)', color='#4A7C40', fontsize=8, labelpad=4)
    ax.tick_params(axis='x', colors='#4A7C40', labelsize=7)

    # Caliper + BS on a twin x (bottom)
    ax_cal = ax.twiny()
    ax_cal.xaxis.set_ticks_position('bottom')
    ax_cal.xaxis.set_label_position('bottom')
    ax_cal.plot(cali,    depth, color='#1A1A1A', lw=0.8, ls='--', label='CALI')
    ax_cal.plot(bs_plot, depth, color='#888888', lw=0.8, ls=':',  label='BS')
    ax_cal.set_xlim(6, 16)
    ax_cal.set_xlabel('CALI / BS (in)', color='#1A1A1A', fontsize=7, labelpad=4)
    ax_cal.tick_params(axis='x', colors='#1A1A1A', labelsize=7)

    ax.set_ylabel('Depth (m MD)', fontsize=10)
    ax.set_title('GR  |  CALI', fontsize=9, fontweight='bold', pad=18)
    ax.grid(True, alpha=0.25, lw=0.5)
    ax.invert_yaxis()

    # ─────────────────────────────────────────────────────────────────────────
    # Track 2 — Resistivity (log scale)
    # ─────────────────────────────────────────────────────────────────────────
    ax = axes[1]
    ax.plot(rt, depth, color='#8B0000', lw=0.8)
    ax.set_xscale('log')
    ax.set_xlim(0.2, 2000)
    ax.xaxis.set_label_position('top')
    ax.xaxis.tick_top()
    ax.set_xlabel('RT (Ω·m)', color='#8B0000', fontsize=8, labelpad=4)
    ax.tick_params(axis='x', colors='#8B0000', labelsize=7)
    ax.set_yticklabels([])
    ax.set_title('RESISTIVITY', fontsize=9, fontweight='bold', pad=18)
    ax.grid(True, alpha=0.25, lw=0.5, which='both')

    # ─────────────────────────────────────────────────────────────────────────
    # Track 3 — Density + Neutron
    # RHOB: left scale (1.95–2.95 g/cc), red
    # NPHI: right scale reversed (0.45–-0.15 v/v), blue
    # Gas crossover: where NPHI (converted to same axis) is left of RHOB
    # ─────────────────────────────────────────────────────────────────────────
    ax = axes[2]
    RHOB_MIN, RHOB_MAX = 1.95, 2.95
    NPHI_MAX, NPHI_MIN = 0.45, -0.15  # axis goes high→low (reversed)

    ax.plot(rhob, depth, color='#C0392B', lw=0.9, label='RHOB')
    ax.set_xlim(RHOB_MIN, RHOB_MAX)
    ax.xaxis.set_label_position('top')
    ax.xaxis.tick_top()
    ax.set_xlabel('RHOB (g/cc)', color='#C0392B', fontsize=8, labelpad=4)
    ax.tick_params(axis='x', colors='#C0392B', labelsize=7)

    # NPHI on twin x (bottom, reversed scale)
    ax_nphi = ax.twiny()
    ax_nphi.xaxis.set_ticks_position('bottom')
    ax_nphi.xaxis.set_label_position('bottom')
    ax_nphi.plot(nphi, depth, color='#2980B9', lw=0.9, label='NPHI')
    ax_nphi.set_xlim(NPHI_MAX, NPHI_MIN)  # reversed: high on left
    ax_nphi.set_xlabel('NPHI (v/v ←)', color='#2980B9', fontsize=7, labelpad=4)
    ax_nphi.tick_params(axis='x', colors='#2980B9', labelsize=7)

    # Gas crossover fill — rescale NPHI to RHOB axis space for fill
    # Map: NPHI_MAX→RHOB_MIN, NPHI_MIN→RHOB_MAX (both axes span same track)
    nphi_on_rhob_axis = RHOB_MIN + (NPHI_MAX - nphi) / (NPHI_MAX - NPHI_MIN) * (RHOB_MAX - RHOB_MIN)
    ax.fill_betweenx(depth, rhob, nphi_on_rhob_axis,
                     where=(nphi_on_rhob_axis < rhob),
                     color='#85C1E9', alpha=0.35, label='Gas crossover')

    ax.set_yticklabels([])
    ax.set_title('RHOB  |  NPHI', fontsize=9, fontweight='bold', pad=18)
    ax.grid(True, alpha=0.25, lw=0.5)

    # ─────────────────────────────────────────────────────────────────────────
    # Track 4 — Compressional + Shear Sonic
    # Both on reversed scale (fast rock on right)
    # ─────────────────────────────────────────────────────────────────────────
    ax = axes[3]
    ax.plot(dt,  depth, color='#1B2631', lw=0.9, label='DT')
    ax.plot(dts, depth, color='#E74C3C', lw=0.9, label='DTS', alpha=0.85)
    ax.set_xlim(300, 40)  # reversed — slow on left, fast on right
    ax.xaxis.set_label_position('top')
    ax.xaxis.tick_top()
    ax.set_xlabel('DT / DTS (μs/ft  →fast)', fontsize=8, labelpad=4)
    ax.tick_params(axis='x', labelsize=7)
    ax.legend(loc='lower right', fontsize=7, framealpha=0.7)
    ax.set_yticklabels([])
    ax.set_title('DT  |  DTS', fontsize=9, fontweight='bold', pad=18)
    ax.grid(True, alpha=0.25, lw=0.5)

    # ─────────────────────────────────────────────────────────────────────────
    # Track 5 — Density Correction (DRHO)
    # Quality thresholds at ±0.15 g/cc
    # ─────────────────────────────────────────────────────────────────────────
    ax = axes[4]
    DRHO_THRESH = 0.15

    ax.plot(drho, depth, color='#7D3C98', lw=0.8)
    ax.fill_betweenx(depth, drho, 0,
                     where=(np.abs(drho) > DRHO_THRESH),
                     color='#E74C3C', alpha=0.40, label=f'|DRHO| > {DRHO_THRESH}')
    ax.fill_betweenx(depth, drho, 0,
                     where=(np.abs(drho) <= DRHO_THRESH),
                     color='#7D3C98', alpha=0.12)
    ax.axvline(0,             color='#555555', lw=0.7)
    ax.axvline( DRHO_THRESH,  color='#E74C3C', lw=0.8, ls='--', label=f'+{DRHO_THRESH}')
    ax.axvline(-DRHO_THRESH,  color='#E74C3C', lw=0.8, ls='--', label=f'-{DRHO_THRESH}')
    ax.set_xlim(-0.5, 0.5)
    ax.xaxis.set_label_position('top')
    ax.xaxis.tick_top()
    ax.set_xlabel('DRHO (g/cc)', color='#7D3C98', fontsize=8, labelpad=4)
    ax.tick_params(axis='x', colors='#7D3C98', labelsize=7)
    ax.legend(loc='lower right', fontsize=6, framealpha=0.7)
    ax.set_yticklabels([])
    ax.set_title('DRHO', fontsize=9, fontweight='bold', pad=18)
    ax.grid(True, alpha=0.25, lw=0.5)

    # ── Figure title and depth axis ───────────────────────────────────────────
    fig.suptitle(
        f'{well_name}  —  Composite Log Display  ({top:.0f}–{base:.0f} m MD)',
        fontsize=12, fontweight='bold', y=1.01
    )

    # Major y-ticks every 100 m, minor every 50 m
    axes[0].yaxis.set_major_locator(ticker.MultipleLocator(100))
    axes[0].yaxis.set_minor_locator(ticker.MultipleLocator(50))
    axes[0].tick_params(axis='y', which='major', length=5)
    axes[0].tick_params(axis='y', which='minor', length=3)

    plt.tight_layout()
    return fig, axes
```

```python
# ── Plot the full main logging interval ──────────────────────────────────────
fig, axes = plot_composite_log(
    df,
    top       = 2585,
    base      = 3680,
    well_name = '15/9-F-1 A  (Volve)',
    bs_default= 8.5,
)
plt.show()
```

## Step 1.5 — Observations from the Composite Display

Work through each track systematically and note what the display is telling you. The observations below are the kind of notes you would record at this stage:

**Track 1 — GR / Caliper**
- GR shows a cyclical sand-shale pattern consistent with alternating reservoir and seal facies
- Clean sand intervals (GR < 60 API, gold fill) identify the potential reservoir zones
- CALI tracks close to bit size throughout → good hole condition; no major washouts visible
- Where CALI departs from BS, correlate with DRHO track to assess density impact

**Track 2 — Resistivity**
- High RT spikes in clean sand intervals → possible hydrocarbon saturation
- Correlate high-RT + low-GR combinations to identify HC candidates
- Low RT in shales (~1–2 Ω·m) sets the brine baseline; use for Rw estimation

**Track 3 — RHOB / NPHI**
- Blue shading marks intervals where NPHI crosses left of RHOB (gas crossover effect)
- These intervals coincide with high-RT zones on Track 2 → strong HC indicators
- Heavy shales show high NPHI (clay-bound water) and moderate RHOB

**Track 4 — DT / DTS**
- DT (navy) and DTS (red) are both present over the logging interval — this well can serve as the Vs calibration well for the RPM
- Both curves show slower velocities (higher μs/ft) in the shale sections and faster in the sands
- Check for DT cycle skips: any isolated DT spike to values > 120 μs/ft in consolidated rock should be flagged
- Vp/Vs ratio (DTS/DT) should be inspected formation by formation in later phases

**Track 5 — DRHO**
- DRHO stays within ±0.15 g/cc for most of the well → density log is largely reliable
- Any red-shaded intervals flag where RHOB should be reviewed or replaced
- Note that positive DRHO (pad bridge) and negative DRHO (gas near borehole wall) have different causes

```python
# ── Zoomed display — reservoir interval only ─────────────────────────────────
# Adjust top / base to focus on the main reservoir section of interest
fig, axes = plot_composite_log(
    df,
    top       = 3050,
    base      = 3450,
    well_name = '15/9-F-1 A  (Volve)  — Reservoir Zoom',
    bs_default= 8.5,
    figsize   = (16, 12),
)
plt.show()
```
